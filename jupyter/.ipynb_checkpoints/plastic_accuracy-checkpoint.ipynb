```python
%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import sys, pickle

sys.path.append('D:\georgia_tech\diverters\src')

import materials, apdl, hemj, linearize, stats
from multilinear import MultilinearModel

duplex = materials.Duplex()
aisi_304 = materials.AISI304()
```

```python
def plot_multi_vs_exp(mat, multi, t, ax):
    duplex.plot_stress_strain(t, 100, axis=ax, label="Experimental")
    multi.plot(t, axis=ax, label="Multilinear")
    
models = {n: MultilinearModel(duplex, n) for n in range(2, 11)}

fig, axes = plt.subplots(2, 2)
fig.suptitle("Stress vs Strain (Multilinear) T=550°C")

T = 550
plot_multi_vs_exp(duplex, models[2], T, axes[0][0])
plot_multi_vs_exp(duplex, models[3], T, axes[0][1])
plot_multi_vs_exp(duplex, models[4], T, axes[1][0])
plot_multi_vs_exp(duplex, models[5], T, axes[1][1])

axes[0][0].set_ylabel("MPa")
axes[1][1].set_xlabel("%")
axes[1][1].legend()

plt.show()
```

```python
def multi_rms_err(model, material, t):
    pts = model.temp_table[t]
    multi_strain = [pt.strain for pt in pts]
    multi_stress = [pt.stress for pt in pts]
    
    strain_sample, stress_sample = stats.normalize(multi_strain[:-1], multi_stress[:-1], 100)
    strain_pred, stress_pred = material.get_stress_strain(t, 100)
    stress_sample = [stress / 1e6 for stress in stress_sample]
    return stats.rms_err(strain_sample, stress_sample, strain_pred, stress_pred)
    
pt_counts = list(range(2, 11))
rms_errs = [multi_rms_err(models[i], duplex, 550) for i in pt_counts]
ax = plt.subplot()
ax.plot(pt_counts, rms_errs)
ax.scatter(pt_counts, rms_errs)
ax.set_title("RMS Error vs Keypoints")
ax.set_xlabel("Keypoints")
ax.set_ylabel("RMS Error (MPa)")
plt.show()
```

```python
def eval_stress(i, line_nodes):
    with open(f"D:\\georgia_tech\\diverters\\src\\results\\hemj\\result_{i}.pkl", "rb") as f:
        result = pickle.load(f)

    node_data.sort(key=lambda x: x[1])
    nodes_x = [data[1] for data in node_data]
    tensors = [result.get_stress_tensor(data[0]) for data in node_data]
    eqv_von_mises = [linearize.get_von_mises(tensor) for tensor in tensors]
    linearized = linearize.linearize(tensors, nodes_x)
    linearized_von_mises = [linearize.get_von_mises(tensor) for tensor in linearized]
    
    center_node = round(len(linearized_von_mises) / 2)
    return eqv_von_mises[center_node], linearized_von_mises[center_node]

node_data = []
with open("D:\\georgia_tech\\diverters\\src\\results\\hemj\\line_nodes.txt") as f:
    for line in f:
        try:
            data = line.split()
            node_data.append((int(data[0]), float(data[1])))
        except ValueError:
            continue

keypoints = list(range(2, 5))
stress_results = [eval_stress(i, node_data) for i in keypoints]
eqv_stresses = [result[0] for result in stress_results]
linear_stresses = [result[1] for result in stress_results]

print(eqv_stresses)
print(linear_stresses)

ax = plt.subplot()
ax.plot(keypoints, eqv_stresses)
ax.scatter(keypoints, eqv_stresses)
# ax.plot(keypoints, linear_stresses)
# ax.scatter(keypoints, linear_stresses)
plt.show()
```
