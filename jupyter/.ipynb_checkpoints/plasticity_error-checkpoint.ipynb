```python
%reload_ext autoreload
%autoreload 2

import matplotlib.pyplot as plt
import numpy as np
import sys, pickle

sys.path.append('D:\georgia_tech\diverters\src')

import materials, hemj, stats
import parametric_solver.apdl_result
import parametric_solver.linearization as linearize
from multilinear import MultilinearModel


duplex = materials.Duplex()
aisi_304 = materials.AISI304()
```

```python
def plot_multi_vs_exp(mat, multi, t, ax):
    duplex.plot_stress_strain(t, 100, axis=ax, label="Experimental")
    multi.plot(t, axis=ax, label="Multilinear")
    
models = {n: MultilinearModel(duplex, n) for n in range(2, 15)}

T = 870

fig, axes = plt.subplots(2, 2)
fig.suptitle(f"Experimental/Mulilinear Comparison\nDuplex {T}C")

plot_multi_vs_exp(duplex, models[2], T, axes[0][0])
plot_multi_vs_exp(duplex, models[3], T, axes[0][1])
plot_multi_vs_exp(duplex, models[4], T, axes[1][0])
plot_multi_vs_exp(duplex, models[5], T, axes[1][1])

axes[0][0].set_ylabel("MPa")
axes[1][1].set_xlabel("%")
axes[1][1].legend()

plt.show()
```

```python
def multi_rms_err(model, material, t):
    pts = model.temp_table[t]
    multi_strain = [pt.strain for pt in pts]
    multi_stress = [pt.stress for pt in pts]
    
    strain_sample, stress_sample = stats.normalize(multi_strain[:-1], multi_stress[:-1], 100)
    strain_pred, stress_pred = material.get_stress_strain(t, 100)
    stress_sample = [stress / 1e6 for stress in stress_sample]
    return stats.rms_err(strain_sample, stress_sample, strain_pred, stress_pred)
    
pt_counts = list(range(2, 15))
rms_errs = [multi_rms_err(models[i], duplex, 870) for i in pt_counts]
ax = plt.subplot()
ax.plot(pt_counts, rms_errs)
ax.scatter(pt_counts, rms_errs)
ax.set_title("RMS Error vs Keypoints")
ax.set_xlabel("Keypoints")
ax.set_ylabel("RMS Error (MPa)")
plt.show()
```

```python
def eval_total_strain(i):
    with open(f"D:\\georgia_tech\\diverters\\src\\results\\hemj_v2\\result_{i}.pkl", "rb") as f:
        result = pickle.load(f)

    total_strains = [result.get_total_strain(data[0]) for data in node_data]
    return total_strains[1]

def eval_linear_strain(i):
    with open(f"D:\\georgia_tech\\diverters\\src\\results\\hemj_v2\\result_{i}.pkl", "rb") as f:
        result = pickle.load(f)

    nodes = [data[0] for data in node_data]
    positions = [(data[1], data[2], data[3]) for data in node_data]
    linear_strains = result.get_linear_strains(nodes, positions)

    linear_von_mises = [linearize.get_von_mises(tensor) for tensor in linear_strains]
    return linear_von_mises[1]

def eval_eqv_stress(i):
    with open(f"D:\\georgia_tech\\diverters\\src\\results\\hemj_v2\\result_{i}.pkl", "rb") as f:
        result = pickle.load(f)
        
    eqv_stresses = [result.get_eqv_stress(data[0]) for data in node_data]
    return eqv_stresses[1]

def eval_linear_stress(i):
    with open(f"D:\\georgia_tech\\diverters\\src\\results\\hemj_v2\\result_{i}.pkl", "rb") as f:
        result = pickle.load(f)

    nodes = [data[0] for data in node_data]
    positions = [(data[1], data[2], data[3]) for data in node_data]
    linear_stresses = result.get_linear_stresses(nodes, positions)
    linear_von_mises = [linearize.get_von_mises(tensor) for tensor in linear_stresses]
    return linear_von_mises[1]

def read_node_data():
    node_data = []
    with open("D:\\georgia_tech\\diverters\\src\\results\\hemj_v2\\line_nodes_side.txt") as f:
        for line in f:
            try:
                data = line.split()
                node_data.append((int(data[0]), float(data[1]), float(data[2]), float(data[3])))
            except ValueError:
                continue
    node_data.sort(key=lambda x: x[3])
    return node_data
```
