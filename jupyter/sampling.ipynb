```python
sampler = PropertySampler()
sampler.add_property("Young's Modulus (MPa)", 1e5, 3e5)
sampler.add_property("Tangent Modulus (MPa)", 2e3, 6e3)
sampler.add_property("Coefficient of Thermal Expansion (1/C)", 3e-6, 7e-6)
sample_df = sampler.random(10)
sample_df
```

```python
def plot_sample(_sample_df):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.scatter(_sample_df.iloc[:, 0], _sample_df.iloc[:, 1], _sample_df.iloc[:, 2])

    col_names = _sample_df.columns.tolist()
    ax.set_xlabel(col_names[0])
    ax.set_ylabel(col_names[1])
    ax.set_zlabel(col_names[2], labelpad=-2)

    plt.subplots_adjust(left=0.0, right=1.20, bottom=0.0, top=1.20)

    plt.show()

plot_sample(sampler.random(10))
plot_sample(sampler.random(100))
plot_sample(sampler.random(1000))
```
